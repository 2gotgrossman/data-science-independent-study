```python
import pandas as pd
import json
import ast
```

```python
df = pd.read_csv("data/NYT_data.csv")
```

```python
df.shape
```

```python
df.head()
```

```python
del df['Unnamed: 0']
del df['Unnamed: 0.1']
del df['Unnamed: 0.1.1']
```

```python
del df['abstract']
del df['blog']
del df['keywords']
del df['score']
del df['section_name']
del df['slideshow_credits']
df = df.rename(columns={'_id':'id'})
del df['multimedia']
```

```python
def byline_cleanup(byline):
    if type(byline) is not type("string"):
        return None
    for key, value in ast.literal_eval(byline).items():
        if key == 'original':
            return value
    else:
        return None
byline = list(map(byline_cleanup, df.byline))
df['byline'] = byline
```

```python
df.headline[1]
```

```python
def headline_cleanup(headline):
    if type(headline) is not type("string"):
        return None
    for key, value in ast.literal_eval(headline).items():
        if key == 'main':
            return value
    else:
        return None
headline = list(map(headline_cleanup, df.headline))
```

```python
headline[:5]
```

```python
df['headline'] = headline
```

```python
df.head()
```

```python
del df['print_page']
```

```python
del df['new_desk']
```

```python
df.to_csv("data/NYT_data_cleaned.csv")
```
