```python
categories = ['Arts', 'Business', 'Obituaries', 'Sports', 'World']
data = defaultdict(pd.DataFrame)
for cat in categories[1:]:
    params['fq'] = "section_name:" + cat
    for i in range(2000//10 ):
        time.sleep(1)
        params['page'] = i
        req = requests.request('GET', url=url, params=params)
        if req.status_code == 200:
            j = req.json()['response']['docs']
            data[cat] = data[cat].append(j)
        else:
            print('ERROR:', req.status_code)
        print(cat, i, data[cat].shape)
            
```
